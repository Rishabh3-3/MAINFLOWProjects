```python
df_cleaned = df.copy()
df_cleaned["index"].fillna(df_cleaned["index"].mean(), inplace=True)
```

```python
df = df.drop_duplicates()
```

```python
df.shape
```

```python
#Outlier detection and removal
```

```python
Q1 = df["Revenue"].quantile(0.25)
Q3 = df["Revenue"].quantile(0.75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5*IQR
upper_bound = Q3 + 1.5*IQR

outliers = df[(df["Revenue"] < lower_bound) | (df["Revenue"] > upper_bound)]

print('Outliers : ', outliers)
```

```python
#Remove outliers

df = df[(df["Revenue"] > lower_bound) & (df["Revenue"] < upper_bound)]
```

```python
#Visualization of outliers

import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(6,4))
sns.boxplot(x=df["Revenue"])
plt.title("Outliers in Revenue")
plt.show()
```

### Summary Report: Exploratory Data Analysis of Student Performance Dataset

## 1. Overview

The sales dataset contains transactional records of product sales including customer demographics, product categories, geographical information, and financial metrics such as unit cost, unit price, total cost, and revenue. Exploratory Data Analysis (EDA) was conducted to understand the structure of the dataset, identify key sales patterns, detect anomalies, and examine relationships between variables.

## 2. Data Cleaning

Before performing the analysis, the dataset was cleaned to ensure data quality and consistency.

Missing values were checked across all columns and handled appropriately using suitable methods such as filling with statistical measures or removing irrelevant columns.
Duplicate records were identified and removed to maintain the integrity of the dataset.

Outliers in numerical variables such as revenue, quantity, and pricing were detected using statistical techniques such as the Interquartile Range (IQR) method.
Irrelevant or incomplete columns were reviewed and handled to avoid distortion in the analysis.

These steps ensured that the dataset used for analysis was reliable and suitable for generating meaningful insights.

## 3. Statistical Analysis

Statistical measures such as mean, median, standard deviation, and variance were used to analyze the distribution of numerical variables including quantity sold, unit cost, unit price, and revenue.

The statistical summary showed variation in product prices and sales quantities, indicating differences in product categories and purchasing behavior. Revenue values varied significantly across transactions, suggesting that some product categories contribute more substantially to overall sales.

## 4. Data Visualization Insights

Sales Distribution

Histograms were used to examine the distribution of numerical variables such as revenue and quantity sold. The results indicated that most transactions generate moderate revenue values, while a smaller number of transactions contribute significantly higher sales.

Outlier Detection

Boxplots were used to detect extreme values in variables such as revenue and unit price. A few transactions showed unusually high revenue compared to the majority, indicating either bulk purchases or high-value product sales.

Correlation Analysis

A correlation heatmap was generated to analyze relationships between numerical variables.

Key observations include:

Revenue showed a strong relationship with quantity sold and unit price, indicating that both the number of units sold and product pricing significantly impact total sales.

Unit cost and unit price showed a positive relationship, reflecting typical pricing strategies where selling price is higher than product cost.

## 5. Trends and Patterns

Several notable trends were identified during the analysis:

Sales activity varies across different product categories and sub-categories.
Certain product categories generate higher revenue due to higher unit prices or larger sales quantities.

Geographic regions such as countries and states contribute differently to overall sales performance.
Customer demographics, including age and gender, may influence purchasing behavior across product categories.

These patterns provide insight into which products and markets contribute most to overall revenue.

## 6. Anomalies

Some anomalies were observed in the dataset:

A small number of transactions recorded significantly higher revenue values compared to the average.
Certain entries contained missing or incomplete values that required cleaning.

Some variables showed unusually high variance, suggesting differences in product pricing and purchase quantities.
These anomalies were reviewed during the data cleaning process to ensure accurate analysis.

## Conclusion

The exploratory analysis of the sales dataset revealed important insights into customer purchasing patterns, product performance, and revenue generation. Sales are influenced primarily by product pricing and quantity sold, while demographic and geographic factors also contribute to variations in purchasing behavior. The findings from this analysis provide a useful foundation for further business analysis, forecasting, and decision-making processes.
